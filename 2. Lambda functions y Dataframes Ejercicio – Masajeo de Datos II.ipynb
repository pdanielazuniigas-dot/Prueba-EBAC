```python
import pandas as pd
import numpy as np

# --- Columnas esperadas (exactas, como en tu dataset) ---
EXPECTED_COLS = [
    "ID","Name","Age","Nationality","Overall","Potential","Club","Value","Wage",
    "Preferred Foot","International Reputation","Skill Moves","Position",
    "Joined","Contract Valid Until","Height","Weight","Release Clause"
]

# 0) Verificación suave (no truena si falta alguna, solo avisa)
missing = [c for c in EXPECTED_COLS if c not in df.columns]
if missing:
    print("⚠️ Estas columnas no están en tu df y se ignorarán:", missing)

# --- 1) Construir “molde” por columna con defaults coherentes ---
molde = {}
for c in df.columns:
    s = df[c].dropna()
    if s.empty:
        molde[c] = np.nan
    else:
        if pd.api.types.is_numeric_dtype(s):
            molde[c] = float(s.median())
        else:
            try:
                molde[c] = s.mode(dropna=True).iloc[0]
            except Exception:
                molde[c] = s.iloc[0]

def safe_set(d, col, val):
    if col in df.columns:
        d[col] = val

# --- 2) ID siguiente (si es numérico) ---
def next_ids(n=2):
    if "ID" in df.columns:
        if pd.api.types.is_numeric_dtype(df["ID"]):
            base = int(pd.to_numeric(df["ID"], errors="coerce").max() or 0)
            return [base + i for i in range(1, n+1)]
        else:
            # Si ID no es numérico, crea IDs string únicos
            return [f"NEW_{i}" for i in range(1, n+1)]
    return [None]*n

id1, id2 = next_ids(2)

# --- 3) Jugador 1: Fernando José Zuñiga Santos (México) ---
p1 = molde.copy()
safe_set(p1, "ID", id1)
safe_set(p1, "Name", "Fernando José Zuñiga Santos")
safe_set(p1, "Nationality", "Mexico")
safe_set(p1, "Age", 24)
safe_set(p1, "Overall", 82)
safe_set(p1, "Potential", 95)
safe_set(p1, "Club", "Analítica FC")
safe_set(p1, "Position", "ST")
safe_set(p1, "Preferred Foot", "Right")
safe_set(p1, "International Reputation", 3)
safe_set(p1, "Skill Moves", 4)
# Estas tres columnas a veces son strings con símbolo/abreviaturas; si en tu df son numéricas, se castea abajo
safe_set(p1, "Value", "€28M")
safe_set(p1, "Wage", "€65K")
safe_set(p1, "Release Clause", "€56M")
# Fechas: respeta el formato de tu df (string tipo "2019-07-01" o solo año). Usamos los defaults si tu df usa otro formato:
safe_set(p1, "Joined", "2023-07-01")
safe_set(p1, "Contract Valid Until", 2028)
# Altura/Peso: respeta el formato del df (algunos usan "5'11\"" y "176lbs"; otros usan numérico). Dejamos valores razonables:
safe_set(p1, "Height", molde.get("Height", "178cm"))
safe_set(p1, "Weight", molde.get("Weight", "73kg"))

# --- 4) Jugador 2: Satiago Quiroga Garcia (México) ---
p2 = molde.copy()
safe_set(p2, "ID", id2)
safe_set(p2, "Name", "Satiago Quiroga Garcia")  # (si querías "Santiago", dime y lo cambio)
safe_set(p2, "Nationality", "Mexico")
safe_set(p2, "Age", 21)
safe_set(p2, "Overall", 76)
safe_set(p2, "Potential", 85)
safe_set(p2, "Club", "Club Deportivo Datos")
safe_set(p2, "Position", "CM")
safe_set(p2, "Preferred Foot", "Left")
safe_set(p2, "International Reputation", 2)
safe_set(p2, "Skill Moves", 3)
safe_set(p2, "Value", "€12M")
safe_set(p2, "Wage", "€25K")
safe_set(p2, "Release Clause", "€30M")
safe_set(p2, "Joined", "2024-01-15")
safe_set(p2, "Contract Valid Until", 2029)
safe_set(p2, "Height", molde.get("Height", "175cm"))
safe_set(p2, "Weight", molde.get("Weight", "70kg"))

# --- 5) Armar DF nuevo con mismas columnas (en el orden del df original) ---
new_players = pd.DataFrame([p1, p2], columns=df.columns)

# --- 6) Intentar castear columnas numéricas a número, sin romper strings monetarios ---
for c in new_players.columns:
    if c in df.columns and pd.api.types.is_numeric_dtype(df[c]):
        new_players[c] = pd.to_numeric(new_players[c], errors="coerce")

# --- 7) Concatenar ---
df = pd.concat([df, new_players], ignore_index=True)

# --- 8) Mostrar verificación (usa nombres EXACTOS de tus columnas) ---
cols_show = [c for c in [
    "ID","Name","Nationality","Age","Overall","Potential","Club","Position","Value","Wage","Release Clause"
] if c in df.columns]

print("✅ Se agregaron jugadores. Vista rápida:")
print(df.tail(5)[cols_show])
```

```python
# Si te falta pycountry, primero ejecuta esta línea (descoméntala):
# %pip install pycountry

import pandas as pd
import pytz
import pycountry
from datetime import datetime

# 1) Cargar tu archivo (ajusta la ruta si cambia)
df = pd.read_csv(r"C:\Users\pdani\Downloads\fifa_eda.csv")

# 2) Detectar la columna de país
country_col = "Nationality" if "Nationality" in df.columns else ("País" if "País" in df.columns else None)
if country_col is None:
    raise KeyError("No encontré la columna de país. Usa 'Nationality' o 'País'.")

# 3) Alias para nombres no estándar
aliases = {
    "England": "United Kingdom",
    "Scotland": "United Kingdom",
    "Wales": "United Kingdom",
    "Northern Ireland": "United Kingdom",
    "Ivory Coast": "Côte d'Ivoire",
    "DR Congo": "Congo, The Democratic Republic of the",
    "Korea Republic": "Korea, Republic of",
    "Korea DPR": "Korea, Democratic People's Republic of",
    "Russia": "Russian Federation",
    "Syria": "Syrian Arab Republic",
    "Iran": "Iran, Islamic Republic of",
    "Tanzania": "Tanzania, United Republic of",
    "Moldova": "Moldova, Republic of",
    "Bolivia": "Bolivia, Plurinational State of",
    "Venezuela": "Venezuela, Bolivarian Republic of",
    "Cape Verde": "Cabo Verde",
    "Curacao": "Curaçao",
    "FYR Macedonia": "North Macedonia",
    "Swaziland": "Eswatini",
    "USA": "United States",
    "U.S.": "United States",
}

# 4) Elegir una zona "típica" para países con varias
prefer_tz = {
    "US": "America/New_York",
    "MX": "America/Mexico_City",
    "BR": "America/Sao_Paulo",
    "CA": "America/Toronto",
    "AU": "Australia/Sydney",
    "AR": "America/Argentina/Buenos_Aires",
    "ES": "Europe/Madrid",
    "GB": "Europe/London",
    "FR": "Europe/Paris",
    "DE": "Europe/Berlin",
    "IT": "Europe/Rome",
    "PT": "Europe/Lisbon",
}

# 5) Funciones básicas
def to_iso2(country_name: str):
    if not isinstance(country_name, str):
        return None
    name = aliases.get(country_name.strip(), country_name.strip())
    try:
        return pycountry.countries.lookup(name).alpha_2
    except:
        return None

def country_to_tzname(country_name: str):
    iso = to_iso2(country_name)
    if iso is None:
        return "Desconocida"
    if iso in prefer_tz:
        return prefer_tz[iso]
    tzs = pytz.country_timezones.get(iso, [])
    return tzs[0] if tzs else "Desconocida"

def tz_to_gmt(tzname: str):
    if tzname == "Desconocida":
        return "Desconocida"
    tz = pytz.timezone(tzname)
    h = int(datetime.now(tz).utcoffset().total_seconds() // 3600)
    return f"GMT{h:+d}"

# 6) Crear columnas
df["TZ Name"]   = df[country_col].apply(country_to_tzname)  # p.ej. Europe/Madrid
df["Time Zone"] = df["TZ Name"].apply(tz_to_gmt)            # p.ej. GMT+1

# 7) Verificar y listar los pocos que falten
print(df[[country_col, "TZ Name", "Time Zone"]].head(12))

faltan = df.loc[df["TZ Name"]=="Desconocida", country_col].dropna().unique()
print("Países sin TZ automática:", faltan)
```

```python

extra_aliases = {
    "Bosnia Herzegovina": "Bosnia and Herzegovina",
    "Central African Rep.": "Central African Republic",
    "Republic of Ireland": "Ireland",
    "China PR": "China",
    "Guinea Bissau": "Guinea-Bissau",
    "Trinidad & Tobago": "Trinidad and Tobago",
    "São Tomé & Príncipe": "Sao Tome and Principe",
    "St Kitts Nevis": "Saint Kitts and Nevis",
    "Antigua & Barbuda": "Antigua and Barbuda",
    "St Lucia": "Saint Lucia",
    "Turkey": "Türkiye",
    "Palestine": "Palestine, State of",
}
# Suma estos alias a los que ya tenías
aliases.update(extra_aliases)

# === 2) Algunos no tienen ISO oficial (ej. Kosovo). Asigna TZ directa por nombre. ===
direct_tz_by_name = {
    "Kosovo": "Europe/Belgrade",  # IANA no tiene 'Europe/Pristina'; usa el huso de Serbia.
}

# === 3) Redefine la función para que primero pruebe el mapeo directo por nombre ===
def country_to_tzname(country_name: str):
    if not isinstance(country_name, str):
        return "Desconocida"
    # Si hay TZ directa por nombre (p.ej. Kosovo), úsala
    if country_name in direct_tz_by_name:
        return direct_tz_by_name[country_name]

    # Normaliza nombre con alias y resuelve ISO
    name_norm = aliases.get(country_name, country_name)
    iso = to_iso2(name_norm)
    if iso is None:
        return "Desconocida"
    if iso in prefer_tz:
        return prefer_tz[iso]

    tzs = pytz.country_timezones.get(iso, [])
    return tzs[0] if tzs else "Desconocida"

# === 4) Regenera columnas de zona horaria ===
df["TZ Name"]   = df[country_col].apply(country_to_tzname)
df["Time Zone"] = df["TZ Name"].apply(tz_to_gmt)

# === 5) Revisa si aún queda alguno sin asignar (deberían ser muy pocos o ninguno) ===
faltan = df.loc[df["TZ Name"]=="Desconocida", country_col].dropna().unique()
print("Países sin TZ automática:", faltan)

# Vista rápida
print(df[[country_col, "TZ Name", "Time Zone"]].head(15))
```
